```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options 
from bs4 import BeautifulSoup
```

```python
options = Options() 
chrome_path = '/Users/carsonlloyd/Downloads/chromedriver'
```

```python
options.add_argument('--window-size=500,300')
```

```python
driver = webdriver.Chrome(chrome_path, 
                              options=options)

driver.set_window_size(1400,1000)
```

```python
driver.get('https://finance.yahoo.com/quote/ATRX/history?period1=755222400&period2=1578787200&interval=1d&filter=history&frequency=1d')
```

```python
import pandas as pd
page_source = driver.page_source
soup = BeautifulSoup(page_source)
tables = soup.findAll('table')
table = tables[0].prettify()
df = pd.read_html(table)[0]
```
